##### Prepare news headlines data in Daily Financial News for 6000+ Stocks Dataset
1. Loại bỏ các dữ liệu thiếu dữ kiện (Null, Nan) 
2. Lấy dữ liệu từ năm 2014 đến hết 2015
3. Lọc các mã có số lượng news headlines thấp

```python
df = pd.read_csv('Datasets/analyst_ratings_processed.csv',index_col = 0)
df.head()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
pre_df = df[df['date'].notna()]
pre_df.isnull().sum()
```

```python
pre_df = df[df['stock'].notna()]
pre_df.isnull().sum()
```

```python
pre_df['year'] = [int(str(d).split(' ')[0].split('-')[0]) for d in pre_df['date']]
pre_df.head()
```

```python
# Lọc news headlines trong thời gian từ 2014 đến 2015
pre_df = pre_df[pre_df.year >= 2014]
pre_df
```

```python
# Lọc news headlines trong thời gian từ 2014 đến 2015
pre_df = pre_df[pre_df.year <= 2015]
```

```python
pre_df
```

```python
pre_df.groupby('year').count()
```

```python
# Lọc các mã có news headline < 25
s = pre_df.stock.value_counts()
pre_df = pre_df[pre_df.stock.isin(s[s >= 25].index)]
```

```python
#Split Datetime để merge với bảng stock price dự trên date time
pre_df['merged'] = [str(d).split(' ')[0] for d in pre_df['date']]
pre_df
```

##### Prepare stock price data in Huge Stock Market Dataset datasets and concat with news headlines data
1. Load data
2. Concat with news headlines dựa trên các tiêu chí cùng stock, datetime

```python
# Load file path and merge 2 bang news headlines va stock price
import os

path = 'Datasets/Stocks/'
arr = []
output = []
stock_names = pre_df.stock.value_counts().index
# print(stock_names)
for file in os.listdir(path) :
    if file.split('.')[0].upper() in stock_names: 
        arr.append([file.split('.')[0],path+file])
        df_stock = pd.read_csv(path+file)
        df_headlines = pre_df[pre_df.stock == file.split('.')[0].upper()]
        merged_df = pd.merge(df_headlines,df_stock, left_on = 'merged',right_on='Date')
        output.append(merged_df)
```

```python
data = pd.concat(output,axis =0, ignore_index =True)
data
```

```python
data['Trend'] = data.Close - data.Open
data['Trend'] = np.where(data.Trend>0,1,0)
data
```

```python
data.title.replace("[^a-zA-Z0-9]"," ",regex=True, inplace=True)
```

```python
data.to_csv("Data_merged.csv")
```
